```sql
/*Affichez le nombre total de commandes (SalesOrderID) effectuÃ©es
par chaque client. (Tables : SalesOrderHeader, Customer)

remarque : en partant du principe que certains clients n'ont pas encore fait de commandes
*/

SELECT 
  C.customerID, 
  C.FirstName, 
  C.Lastname, 
  count(soh.SalesOrderID) as TotalOrders  
FROM saleslt.customer as C
LEFT JOIN salesorderheader as SOH
on C.CustomerID = SOH.CustomerID
GROUP BY C.customerID, C.FirstName, C.Lastname
--ajouter "HAVING totalorders > 0" pour exclure les clients qui n'ont pas encore fait de commandes
SORT BY totalorders DESC ;

```
