```sql
/* Produits par catégorie :
Affichez le nombre total de produits dans chaque catégorie de produit.
(Tables : Product, ProductCategory) */

select 
  pc.Name as ProductCategoryName, 
  count(p.ProductCategoryID) as NbProducts
from productcategory as pc 
  left join product as p on
  p.ProductCategoryID = pc.ProductCategoryID
group by pc.Name
order by NbProducts desc ;
```
