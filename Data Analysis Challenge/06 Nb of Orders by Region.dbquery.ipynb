```sql
/*Commandes par région : Trouvez le nombre de commandes
pour chaque région (StateProvince), triées par région.
(Tables : Address, SalesOrderHeader) */

select 
  a.StateProvince as `Region`, 
  count(soh.SalesOrderID) as NbOrders
from Address as A 
 left join SalesOrderHeader as SOH
on soh.ShipToAddressID = a.AddressID
group by StateProvince
order by StateProvince ;
```
