```sql
/*
Analyse des remises :
Calculez le total des remises (UnitPriceDiscount) appliquées par
produit et identifiez celui avec la remise totale la plus élevée.
(Table : SalesOrderDetail)
Remarque : le total des remises par produit se trouve dans la colonne TotalProductDiscounts
*/

select 
 ProductID,
 sum(OrderQty) as TotalQuantity,
 sum(orderqty * UnitPrice) as TotalPrice,
 sum(LineTotal) as TotalRevenue,
 sum((OrderQty * UnitPrice) - LineTotal) as TotalProductDiscounts
from salesorderDetail
where UnitPriceDiscount > 0
group by ProductID 
order by totalproductdiscounts desc
limit 1;
```
