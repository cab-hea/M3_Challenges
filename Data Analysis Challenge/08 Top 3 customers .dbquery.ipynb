```sql
/* Top 3 clients générant le plus de revenus :
Identifiez les 3 clients ayant généré le plus de revenus totaux à travers leurs
commandes.
(Tables : SalesOrderHeader, Customer) */

Select 
 c.CustomerID,
 c.FirstName,
 c.LastName,
 sum(SOH.TotalDue) as Revenue
from customer as C
left join salesorderheader as SOH
on C.CustomerID = SOH.CustomerID
group by c.CustomerID, c.FirstName, c.LastName
order by Revenue desc
limit 3;
```
