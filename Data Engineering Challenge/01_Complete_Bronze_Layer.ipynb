# Complete Bronze Layer
Récupérez dans Databricks toutes les tables de la base de données AdventureWorksLT, sauf les tables de metadata

### Connecting to the bronze layer (target)

```sql
USE CATALOG cab_lakehouse;
USE DATABASE bronze;
```

### Load data into bronze layer of Lakehouse
Tables à charger : 
CustomerAddress, ProductModel, ProductModelProductDescription, ProductDescription

```sql
%sql
-- Ingestion of CustomerAddress
CREATE OR REPLACE TABLE customeraddress
AS SELECT * FROM cab_adventureworks.saleslt.customeraddress;
```

```sql
-- Ingestion of ProductDescription
CREATE OR REPLACE TABLE productdescription
AS SELECT * FROM cab_adventureworks.saleslt.productdescription;
```

```sql

-- Ingestion of ProductModel
CREATE OR REPLACE TABLE productmodel
AS SELECT * FROM cab_adventureworks.saleslt.productmodel;
```

```sql

-- Ingestion of ProductModelProductDescription
CREATE OR REPLACE TABLE productmodelproductdescription
AS SELECT * FROM cab_adventureworks.saleslt.productmodelproductdescription;
```
